```python
#  Import necessary libraries
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

#  Set your dataset path (this matches what we used while unzipping)
dataset_path = '/content/waste_classification'

#  List all waste categories (folder names)
categories = os.listdir(dataset_path)
print("Waste Categories Found:", categories)

#  Count number of images in each category
print("\nNumber of images in each category:\n")
for category in categories:
    folder_path = os.path.join(dataset_path, category)
    if os.path.isdir(folder_path):
        print(f"{category}: {len(os.listdir(folder_path))} images")

#  Show one random image from each category
print("\nDisplaying sample images from each category:")
for category in categories:
    folder_path = os.path.join(dataset_path, category)
    sample_image = random.choice(os.listdir(folder_path))
    image_path = os.path.join(folder_path, sample_image)

    img = Image.open(image_path)
    plt.imshow(img)
    plt.title(category)
    plt.axis('off')
    plt.show()
```
